```python
# Import all good things
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# The constants define the signal, noise, and "instrument" properties
SIG_AMPLITUDE = 10
SIG_OFFSET = 2
SIG_PERIOD = 100
NOISE_AMPLITUDE = 3
N_SAMPLES = 5 * SIG_PERIOD
INSTRUMENT_RANGE = 9

# Construct a sine wave and mix it with some random noise
times = np.arange(N_SAMPLES).astype(float)
signal = SIG_AMPLITUDE * np.sin(2 * np.pi * times / SIG_PERIOD) + SIG_OFFSET
noise = NOISE_AMPLITUDE * np.random.normal(size=N_SAMPLES)
signal += noise

# Truncate spikes that are outside of the instrument range
signal[signal > INSTRUMENT_RANGE] = INSTRUMENT_RANGE
signal[signal < -INSTRUMENT_RANGE] = -INSTRUMENT_RANGE

# Plot the results
matplotlib.style.use("ggplot")
plt.plot(times, signal)
plt.title("Synthetic sine wave signal")
plt.xlabel("Time")
plt.ylabel("Signal + noise")
plt.ylim(ymin=-INSTRUMENT_RANGE, ymax=INSTRUMENT_RANGE)

# Save the plot
plt.savefig("signal.pdf")

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Number of observations
n = 5000

# Create a starting day
day = pd.to_datetime('2013-02-06')

# Generate random seconds spanning 28800 seconds (8 hours)
seconds = np.random.rand(n) * 28800 * pd.Timedelta(1, 's')

# Start time at 8 am
start = pd.offsets.Hour(8)

# Create irregular time series
tidx = day + start + seconds
tidx = pd.Series(np.sort(tidx), name='DateTime')

# Generate the time series data
s = pd.Series(np.random.randn(n), index=tidx, name='A').cumsum()

# Plot the time series
plt.figure(figsize=(12, 6))
s.plot(title='Irregular Time Series with Cumulative Sum')
plt.xlabel('DateTime')
plt.ylabel('Value')
plt.show()
```

```python
import pandas as pd
import numpy as np

# Create DataFrames
df1 = pd.DataFrame({'x': [1, 2, 3], 'y': ['a', 'b', 'c']})
df2 = pd.DataFrame({'y': ['b', 'c', 'd'], 'z': [4, 5, 6]})

# Display the DataFrames
print("DataFrame df1:")
print(df1)
print("\nDataFrame df2:")
print(df2)

# 1. Inner Join (only common rows in both DataFrames)
inner_join_df = pd.merge(df1, df2, on='y', how='inner')
print("\nInner Join DataFrame:")
print(inner_join_df)

# 2. Left Join (all rows from df1, matched rows from df2)
left_join_df = pd.merge(df1, df2, on='y', how='left')
print("\nLeft Join DataFrame:")
print(left_join_df)

# 3. Right Join (all rows from df2, matched rows from df1)
right_join_df = pd.merge(df1, df2, on='y', how='right')
print("\nRight Join DataFrame:")
print(right_join_df)

# 4. Outer Join (all rows from both DataFrames)
outer_join_df = pd.merge(df1, df2, on='y', how='outer')
print("\nOuter Join DataFrame:")
print(outer_join_df)

# 5. Concatenation (rows concatenation)
concat_rows = pd.concat([df1, df2], axis=0, ignore_index=True, sort=False)
print("\nConcatenated DataFrame (rows):")
print(concat_rows)

# 6. Concatenation (columns concatenation)
concat_columns = pd.concat([df1, df2], axis=1)
print("\nConcatenated DataFrame (columns):")
print(concat_columns)
```

```python
import cmath
import matplotlib.pyplot as plt
import numpy as np

# Initialize lists to store x and y coordinates
x = []
y = []

# Compute values
for t in range(0, 2000):
    s = t * 2 * np.pi / 200
    g = cmath.exp(-2j * s) + (1/2) * cmath.exp(5j * s) + (1/5) * cmath.exp(19j * s)
    x.append(g.real)
    y.append(g.imag)

# Plot the results
plt.figure(figsize=(8, 8))
plt.axis('off')
plt.axis('equal')
plt.plot(x, y)
plt.title('Plot of Complex Function')
plt.show()
```

```python
from math import cos, sin, pi, exp
import matplotlib.pyplot as plt

def f(x):
    return 2 * exp(-x**2)

def g(x):
    return cos(sin(x) + cos(x))

# Initialize lists
x = []
fl = []
gl = []

# Compute values
for i in range(0, 100):
    s = -pi + i * 2 * pi / 100
    x.append(s)
    fl.append(f(s))
    gl.append(g(s))

# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(x, fl, label='f(x) = 2 * exp(-x**2)')
plt.plot(x, gl, label='g(x) = cos(sin(x) + cos(x))')

# Add titles and labels
plt.title('Plots of f(x) and g(x)')
plt.xlabel('x')
plt.ylabel('Function values')
plt.legend()

# Show the plot
plt.show()
```
